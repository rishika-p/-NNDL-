```
x = [0.2,0.4,0.6,0.8]
y = [3.4,3.8,4.2,4.6]
m = 1
c = -1
epochs = 100
number_of_samples = 4
learning_rate = 0.1
err_list = []
def gradient_of_c(y,x,m,c):
    return -1*(y-m*x-c)
def gradient_of_m(y,x,m,c):
    return -1*(y-m*x-c)*x
for i in range(epochs):
    for j in range(number_of_samples):
        err = 0.5 * (y[j] - m*x[j] - c)**2
        err_list.append(err)
        delta_m = -1*learning_rate*gradient_of_m(y[j],x[j],m,c)
        delta_c = -1*learning_rate*gradient_of_c(y[j],x[j],m,c)
        m = m + delta_m
        c = c + delta_c
print("value for c and m are: ",c , m)
```

```

x1 = range(epochs*4)
y1 = error_list

plt.plot(x1,y1)
plt.xlabel("number of iterations")
plt.ylabel("error")
plt.title("error graph in case of SGD")
```
